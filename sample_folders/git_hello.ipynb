```python
from pyspark.sql.types import StructType,StructField,IntegerType,ByteType,StringType,DataType,ShortType
schema = StructType([
    StructField("ID",IntegerType(),True),
    StructField("Name",StringType(),True),
    StructField("Sex",StringType(),True),
    StructField("Age",ByteType(),True),
    StructField("Height",ShortType(),True),
    StructField("Weight",ShortType(),True),
    StructField("Team",StringType(),True),
    StructField("NOC",StringType(),True),
    StructField("Games",StringType(),True),
    StructField("Year",ShortType(),True),
    StructField("Season",StringType(),True),
    StructField("City",StringType(),True),
    StructField("Sport",StringType(),True),
    StructField("Event",StringType(),True),
    StructField("Medal",StringType(),True)
])
df= spark.read.csv("/Workspace/Users/dilip.infoetl@gmail.com/ramesh_ratnasamy/db-git-folder/sample_folders/athlete_events.csv",header=True,schema=schema)
```

```python
display(df.take(10))
```

```python
from pyspark.sql.functions import col,sum

df_nulls = df.select([sum(col(c).isNull().cast("int")).alias(c) for c in df.columns])
display(df_nulls)
```

```python
df.groupBy("NOC").count().filter("NOC IN ('SWE','NOR','FIN','DEN','ICE')").show()
```

```python
df.createOrReplaceTempView("data_table")

df_table = spark.sql("""
SELECT GAMES, NOC, SPORT, COUNT(MEDAL) AS TOTAL_MEADELS
FROM data_table
WHERE MEDAL IN ("Gold","Silver","Bronze")
GROUP BY GAMES,NOC,SPORT
ORDER BY GAMES
""")
display(df_table)
```

```python
df_table.plot(kind="bar",x="SPORT",y="TOTAL_MEADELS",title="graphmedals")
```

```python
fig=df_table.plot(kind="bar",x="SPORT",y="TOTAL_MEADELS",title="graphmedals")
fig.update_layout(yaxis={"autorange":"reversed"})
```

```python
%sql
SELECT GAMES, NOC, SPORT, COUNT(MEDAL) AS TOTAL_MEADELS
FROM data_table
WHERE MEDAL IN ("Gold","Silver","Bronze")
GROUP BY GAMES,NOC,SPORT
ORDER BY GAMES
```

```python
%sql
SHOW CATALOGS;
DROP CATALOG IF EXISTS data CASCADE;
CREATE CATALOG IF NOT EXISTS DATA;
CREATE SCHEMA IF NOT EXISTS DATA.OLYMPIC;
CREATE TABLE IF NOT EXISTS DATA.OLYMPIC.SWEDEN
AS
(
    SELECT *FROM data_table
    WHERE NOC='SWE'
);
```
